```python
#to extract the source content (including html tags) from a website.
import requests
page = requests.get("https://en.wikipedia.org/wiki/Web_mining")
#print(page.text)

#the html tags from the content
from bs4 import BeautifulSoup
html = page.text
soup = BeautifulSoup(html,'html.parser')
html_tag = [tag.name for tag in soup.find_all()]
#print(html_tag)#all list used in order
#print(set(html_tag))#set of list used (no repetition of tag)

#filtering the html tags from the content
#removing or extracting useless data
filter_residue = [s.extract() for s in soup(['style', 'script', '[document]', 'head', 'title'])]
#printing useful text
text = soup.get_text()
#print(text)

# simple tokenization
import nltk
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text)
print(tokens)

#tokenizer to accept only alphanumeric character
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(text)
print(tokens)

#applying stopword removal
from nltk.corpus import stopwords
stopwords = list(nltk.corpus.stopwords.words('english'))
words = [word for word in tokens if word not in stopwords]
#print(words)

#display the number of words and term frequency of each word present in 
#them after applying stop word removal.
fdist = nltk.FreqDist(words)
for word, frequency in fdist.most_common():
    print('{}  ==> {}'.format(word, frequency))


#display the number of words and term frequency of each word present
#along with their corresponding stemmed words present
#after applying stemming to whole text
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stem_words=[]
for word in words:
    stem_words.append(stemmer.stem(word))
fdist = nltk.FreqDist(stem_words)
for word, frequency in fdist.most_common():
    print('{}  ==>  {}  ==> {}'.format(word, frequency,stemmer.stem(word)))

#Adding new word to nltk stopword list
newStopWords = ['web','mine','data']
stopwords.extend(newStopWords)
#now again applying stopword removal to remove new word
words = [word for word in stem_words if word not in stopwords]
fdist = nltk.FreqDist(words)
for word, frequency in fdist.most_common():
    print('{}  ==> {}'.format(word, frequency))


```
