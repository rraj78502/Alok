```python
import requests
from nltk.tokenize import word_tokenize
from nltk.tokenize import RegexpTokenizer
import os
from bs4 import BeautifulSoup
import nltk
import functools
import glob
from pprint import pprint as pp
from nltk.corpus import stopwords
tokenizer = RegexpTokenizer(r'\w+')
stopwords = list(nltk.corpus.stopwords.words('english'))
```

```python
n = input('Enter the no of website you want to search => ')
websites=[]
files=[]
for i in range(int(n)):
    link = input(f'Enter the link {i+1}  ==> ')
    websites.append(link)
Query = input('Enter your Query ==> ')

def saving_pages_content(url,file):
    page = requests.get(url)
    html = page.content
    soup = BeautifulSoup(html,'html.parser')
    filter_residue = [s.extract() for s in soup(['style', 'script', '[document]', 'head', 'title'])]
    text = soup.get_text()
    with open(file,"w",encoding='utf-8') as f:
        f.write(text)
for url in websites:
    file = os.getcwd()+os.sep+url.split('.')[1]+'_'+url.split('.')[-1].split('/')[-1]+'.txt'
    files.append(file)
    saving_pages_content(url,file)

def parsetexts(fileglob='*.txt'):
    texts, words = {}, set()
    for txtfile in glob.glob(fileglob):
        with open(txtfile, 'r',encoding='utf-8') as f:
            txt = f.read()
            txt = word_tokenize(txt)
            #txt = [word for word in txt if word.isalpha()]  
            #txt = [w for w in words if not w in stop_words]
            words |= set(txt)
            texts[txtfile.split('\\')[-1]] = txt
            #print(texts)
    return texts, words

def termsearch(terms): # Searches simple inverted index
    return functools.reduce(set.intersection,(invindex[term] for term in terms),set(texts.keys()))
 
texts, words = parsetexts()
#print('\nTexts')
#pp(texts)
#print('\nWords')
#pp(sorted(words))
 
invindex = {word:set(txt
                        for txt, wrds in texts.items() if word in wrds)
            for word in words}
#print('\nInverted Index')
#pp({k:sorted(v) for k,v in invindex.items()})
 
terms = [word for word in Query.split() if word not in stopwords]
#print('\nTerm Search for: ' + repr(terms))
#print(sorted(termsearch(terms)))
```

```python
doc1 = texts['wikipedia_Cat.txt']
doc2 = texts['wikipedia_Dog.txt']
query = terms
wordSet = set(doc1).union(set(doc2))
wordDictA = dict.fromkeys(wordSet, 0) 
wordDictB = dict.fromkeys(wordSet, 0)
wordDictquery = dict.fromkeys(wordSet, 0)
for word in doc1:
    wordDictA[word]+=1
for word in doc2:
    wordDictB[word]+=1
for word in query:
    wordDictquery[word]+=1
```

```python
def computeTF(wordDict, bow):
    tfDict = {}
    bowCount = len(bow)
    for word, count in wordDict.items():
        tfDict[word] = count/float(bowCount)
    return tfDict

tfBowA = computeTF(wordDictA, doc1)
tfBowB = computeTF(wordDictB, doc2)
tfBowA,tfBowB
```

```python
def computeIDF(docList):
    import math
    idfDict = {}
    N = len(docList)
    
    idfDict = dict.fromkeys(docList[0].keys(), 0)
    for doc in docList:
        for word, val in doc.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items():
        idfDict[word] = math.log10(N / float(val))
        
    return idfDict

idfs = computeIDF([wordDictA, wordDictB])
idfs
```

```python
def computeTFIDF(tfBow, idfs):
    tfidf = {}
    for word, val in tfBow.items():
        tfidf[word] = val*idfs[word]
    return tfidf

tfidfBowA = computeTFIDF(tfBowA, idfs)
tfidfBowB = computeTFIDF(tfBowB, idfs)

import pandas as pd
pd.DataFrame([tfidfBowA, tfidfBowB])
```

```python
import numpy as np

def cos_sim(a, b):
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    return dot_product / (norm_a * norm_b)

Cat = np.array(list(wordDictA.values()))
Dog = np.array(list(wordDictB.values()))
q = np.array(list(wordDictquery.values()))

print(cos_sim(Cat,Dog))
print(cos_sim(Cat,q))
print(cos_sim(Dog,q))
```
