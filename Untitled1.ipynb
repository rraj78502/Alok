```python
#a program to extract the contents (excluding the html tags) from two
#educational websites and save the content in two separate .doc file

import requests
from bs4 import BeautifulSoup

page1 = requests.get("https://coursera.org/")
html1 = page1.text
soup1 = BeautifulSoup(html1,'html.parser')
filter_residue1 = [s.extract() for s in soup1(['style', 'script', '[document]', 'head', 'title'])]
text1 = soup1.get_text()
with open("C:/Users/Rabi/Desktop/Alok/coursera.docx","w",encoding='utf-8') as f:
    f.write(text1)


page2 = requests.get("https://www.udemy.com/")
html2 = page2.text
soup2 = BeautifulSoup(html2,'html.parser')
filter_residue2 = [s.extract() for s in soup2(['style', 'script', '[document]', 'head', 'title'])]
text2 = soup2.get_text()
with open("C:/Users/Rabi/Desktop/Alok/udemy.docx","w",encoding='utf-8') as f:
    f.write(text2)




```

```python
#removing stopword from content
import nltk
from nltk.tokenize import word_tokenize
doc1 = word_tokenize(text1)
doc2 = word_tokenize(text2)

from nltk.corpus import stopwords
stopwords = list(nltk.corpus.stopwords.words('english'))
doc1 = [word for word in doc1 if word not in stopwords]
doc2 = [word for word in doc2 if word not in stopwords]

doc1,doc2
```

```python
#creating bag of word and representing the documents using Boolean
wordSet = set(doc1).union(set(doc2))

wordDictA = dict.fromkeys(wordSet, 0) 
wordDictB = dict.fromkeys(wordSet, 0)

for word in doc1:
    wordDictA[word]+=1
    
for word in doc2:
    wordDictB[word]+=1
    
import pandas as pd
pd.DataFrame([wordDictA, wordDictB])
```

```python
def computeTF(wordDict, bow):
    tfDict = {}
    bowCount = len(bow)
    for word, count in wordDict.items():
        tfDict[word] = count/float(bowCount)
    return tfDict

tfBowA = computeTF(wordDictA, doc1)
tfBowB = computeTF(wordDictB, doc2)
tfBowA,tfBowB
```

```python
def computeIDF(docList):
    import math
    idfDict = {}
    N = len(docList)
    
    idfDict = dict.fromkeys(docList[0].keys(), 0)
    for doc in docList:
        for word, val in doc.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items():
        idfDict[word] = math.log10(N / float(val))
        
    return idfDict

idfs = computeIDF([wordDictA, wordDictB])
idfs
```

```python
def computeTFIDF(tfBow, idfs):
    tfidf = {}
    for word, val in tfBow.items():
        tfidf[word] = val*idfs[word]
    return tfidf

tfidfBowA = computeTFIDF(tfBowA, idfs)
tfidfBowB = computeTFIDF(tfBowB, idfs)

import pandas as pd
pd.DataFrame([tfidfBowA, tfidfBowB])
```
